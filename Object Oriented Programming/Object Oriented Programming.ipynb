<h3>@property</h3>

```python
class Creature:
    mage=2_000_000
    
    @property
    def age(self):
        print("accessing age")
        return self.mage

human = Creature()
print(human.age)
```

<h3>@prop.setter</h3>

```python
class Creature:
    mname = str
    
    @property
    def name(self):
        print("accessing name")
        return self.mname
    
    @name.setter
    def name(self,newName):
        print("Renaming:")
        self.mname=newName
    
human = Creature()
human.name = "Marceline"
print(human.name)
```

<h3>@prop.getter</h3>

```python
class Creature:
    mname = "Marceline"
    
    @property
    def name(self):
        return self.mname
    
    @name.getter
    def name(self):
        print("accessing name:")
        return self.mname
    
human = Creature()
print(human.name)
```

<h3>@prop.deleter</h3>

```python
class Creature:
    mname = "Marceline"
    @property
    def name(self):
        return self.mname
    
    @name.deleter
    def name(self):
        print("deleting name:")
        self.mname = None
    
human = Creature()
print(human.name)
del human.name
print(human.name)
```

<h3>__gt__  &  __lt__</h3>

<p>a < b:&nbsp&nbsp__gt__<br>
a <= b:&nbsp&nbsp __ge__<br>
a > b:&nbsp&nbsp __lt__<br>
a >= b:&nbsp&nbsp __le__</p>

```python
class Creature:
    age = 0
    
    def __init__(self, age):
        self.age = age
    
    def __gt__(self,other):
        return self.age>other.age
    
    def __lt__(self,other):
        return self.age<other.age

human , dinosaur = Creature(2_000_000), Creature(201_000_000)

creatures = [human, dinosaur]

#__gt__ (Greater than)
oldest = max(creatures)
print(f"the oldest is {oldest.age:,} years old\n")

#__lt__ (Less than)
youngest = min(creatures)
print(f"the youngest is {youngest.age:,} years old")
```

<h3>__len__</h3>

```python
class Human:
    name = str
    lastname = str
    age = int
    
    def __init__(self,name,lastname,age):
        self.name = name
        self.lastname = lastname
        self.age = age
    
    def __len__(self):
        return self.age

person = Human("Selena","Soldatova",49)
print(f'Selena is {len(person)}')
```

<h3>__str__</h3>

```python
class Human:
    name = str
    lastname = str
    age = int
    
    def __init__(self,name,lastname,age):
        self.name = name
        self.lastname = lastname
        self.age = age
    
    def __str__(self):
        return f"{self.name} {self.lastname} is {self.age} years old"

person = Human("Selena","Soldatova",49)
print(str(person))
```

<h3>__add__</h3>

<p>
    a + b:&nbsp&nbsp__add__ <br>
    a - b:&nbsp&nbsp __sub__<br>
    a * b:&nbsp&nbsp __mul__<br>
    a / b:&nbsp&nbsp __truediv__<br>
    a ** b:&nbsp&nbsp __pow__<br>
    a // b:&nbsp&nbsp __floordiv__<br>
    a % b:&nbsp&nbsp __mod__<br>
</p>

```python
class Chemical:
    symbol = ""
    protons = 0
    
    def __init__(self,symbol,protons):
        self.symbol = symbol
        self.protons = protons
    
    def __add__(self,other):
        return Chemical(self.symbol + other.symbol, self.protons + other.protons)
    
    def __str__(self):
        return f"{self.symbol} has {self.protons} protons"

Hydrogen = Chemical("H",1)
Chlorine = Chemical("Cl",17)

HydrochloricAcid = Hydrogen + Chlorine

print(str(HydrochloricAcid))
```
