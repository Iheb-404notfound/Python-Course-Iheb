```python
from math import *

def sin_lists(start,end,step=0.1):
    xs,ys = [],[]

    index = start
    while index < end :
        xs.append(index)
        ys.append(sin(index))
        index+=step
        
    return xs,ys

xs,ys = sin_lists(0,25)
plt.plot(xs,ys)
plt.show()
```

<h3>figure

```python
fig = plt.figure()
```

<h3>add_subplot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(xs,ys)
plt.show()
```

<h3>subplots

```python
fig , ax = plt.subplots(1,1)
_,ys2 = sin_lists(1.57,26.57)
ax.plot(xs,ys,xs,ys2)
plt.show()
```

<h3>legend

```python
fig , ax = plt.subplots(1,1)
ax.plot(xs,ys,xs,ys2)
ax.legend(["sin(x)","cos(x)"])
plt.show()
```

<h3>let's imitate density wave function

```python
def wave(start,end,extremum_x,extremum_y=1,slope=0.1,step=0.1):
    xs,ys = [],[]
    x = start
    while x<end:
        xs.append(x)
        ys.append(abs(extremum_y/((slope)*(x-extremum_x)**2+1)))
        x+=step
    return xs,ys

fig, ax = plt.subplots(1,1)
xs,ys = wave(0,18,9)
ax.plot(xs,ys)
ax.text(12.5,0.6,r"$\frac{1}{\frac{1}{10}(x-9)^2+1}$",size=20)
plt.show()
```

<h3>fill

```python
fig, ax = plt.subplots(1,1)
xs,ys = wave(0,18,9)

ax.fill(xs,ys,color="#1c9cc6")

ax.text(12.5,0.9,r"$\int{\frac{1}{\frac{1}{10}(x-9)^2+1}}dx$",size=15)
ax.text(12.5,0.7,r"$=tan^{-1}(\frac{x-9}{\sqrt{10}})$",size=13)
plt.show()
```

<h3>fill_between

```python
fig, ax = plt.subplots(1,1)
xs,ys = wave(0,18,9)
ax.fill(xs,ys,color="#1c9cc6")
ax.fill_between(xs[70:110],ys[0],ys[70:110],color="red")

ax.text(11.5,0.9,r"$\int_{7}^{11}{\frac{1}{\frac{1}{10}(x-9)^2+1}}dx$",size=15)
ax.text(12.5,0.7,r"$=1.127$",size=13)
plt.show()
```

```python
fig , ax = plt.subplots(1,1)

xs,ys = wave(0,18,9,0.1)
_,ys2 = wave(0,18,7,0.1)
ax.stackplot(xs,ys,ys2)
plt.show()
```

<h3>pie

```python
fig , ax = plt.subplots(1,1)
labels = ["You","Also you","Also you","People"]
values = [ 54.8  , 25.2       , 19.9       , 0.1      ]
wedges,_,_ = ax.pie(values, labels=labels, explode=(0,0,0,0.7), shadow=True, autopct='%.1f%%')
ax.set_title("Who can help building yourself")
ax.legend(wedges,labels,title="Builders",loc="upper right")
plt.show()
```
