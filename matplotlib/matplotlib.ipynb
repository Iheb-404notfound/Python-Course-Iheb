```python
from math import *

def sin_lists(start,end,step=0.1):
    xs,ys = [],[]

    index = start
    while index < end :
        xs.append(index)
        ys.append(sin(index))
        index+=step
        
    return xs,ys

xs,ys = sin_lists(0,25)
plt.plot(xs,ys)
plt.show()
```

<h3>figure

```python
fig = plt.figure()
```

<h3>add_subplot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(xs,ys)
plt.show()
```

<h3>subplots

```python
fig , ax = plt.subplots(1,1)
_,ys2 = sin_lists(1.57,26.57)
ax.plot(xs,ys,xs,ys2)
plt.show()
```

<h3>legend

```python
fig , ax = plt.subplots(1,1)
ax.plot(xs,ys,xs,ys2)
ax.legend(["sin(x)","cos(x)"])
plt.show()
```

<h3>set_title

```python
fig , ax = plt.subplots(1,1)
ax.plot(xs,ys)
ax.set_title("This is a circular function")
plt.show()
```

<h3>grid

```python
fig , ax = plt.subplots(1,1)
ax.plot(xs,ys)
ax.grid(visible=True,linestyle="--",color="r",alpha=1)
plt.show()
```

<h3>tick_params

```python
fig , ax = plt.subplots(1,1)
ax.plot(xs,ys)
ax.tick_params(size=50, labelsize=20, labelcolor="r", length=10, width=20, colors='b')
plt.show()
```

<h3>set_xlabel set_ylabel

```python
fig , ax = plt.subplots(1,1)
x = ax.plot(xs,ys,label="sin(x)")
ax.set_title("LC")
ax.set_xlabel("time")
ax.set_ylabel("voltage")
plt.show()
```

<h3>pie

```python
fig , ax = plt.subplots(1,1)
labels = ["You","Also you","Also you","People"]
values = [ 54.8  , 25.2       , 19.9       , 0.1      ]
wedges,_,_ = ax.pie(values, labels=labels, explode=(0,0,0,0.7), shadow=True, autopct='%.1f%%')
ax.set_title("Who can help building yourself")
ax.legend(wedges,labels,title="Builders",loc="upper right")
plt.show()
```

<h3>bar

```python
fig , ax = plt.subplots(1,1)
populations = [2_773_019_936, 3_700_437_046 , 4_870_921_740, 6_143_493_823, 7_379_797_139]
years = [ 1955        , 1970          , 1985         , 2000         , 2015         ]
ax.bar(years,populations,width=4)
ax.set_title("The number of people on earth")
plt.show()
```
