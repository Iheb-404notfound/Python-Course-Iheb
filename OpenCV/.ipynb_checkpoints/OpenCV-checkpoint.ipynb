<h3>addWeighted

```python
black_donkey = cv.imread("black_donkey.png",1)
lines = cv.imread("lines.png",1)

white_and_black_zebra = cv.addWeighted(black_donkey,1,lines,1,0)

plt.subplot(1,3,1)
plt.imshow(cv.cvtColor(black_donkey,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(cv.cvtColor(lines,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(cv.cvtColor(white_and_black_zebra,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

<h3>subtract

```python
donkey = cv.imread("donkey.png",1)
lines = cv.imread("lines.png",1)

black_and_white_zebra = cv.subtract(donkey,lines)

plt.subplot(1,3,1)
plt.imshow(cv.cvtColor(donkey,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(cv.cvtColor(lines,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(cv.cvtColor(black_and_white_zebra,cv.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

<h3>bitwise_and

```python
donkey = cv.imread("donkey.png",1)
lines = cv.imread("lines.png",1)

bitwise_and_zebra = cv.bitwise_and(donkey,lines,mask=None)

plt.imshow(cv.cvtColor(bitwise_and_zebra,cv.COLOR_BGR2RGB))
plt.show()
```

<h3>resize

```python
foal = cv.resize(black_and_white_zebra,(300,200))

plt.imshow(cv.cvtColor(foal,cv.COLOR_BGR2RGB))
plt.show()
```

<h3>shape

```python
foal.shape
```

<h3>warpAffine

```python
height = foal.shape[0]
width = foal.shape[1]
rotation_degrees = 60
centerx = width/2
centery = height/2

rotational_transformation_matrix = cv.getRotationMatrix2D((centerx,centery),rotation_degrees,1)

unicorn = cv.warpAffine(foal, rotational_transformation_matrix, (width,height))

plt.imshow(cv.cvtColor(unicorn,cv.COLOR_BGR2RGB))
plt.show()
```

<h3>adptiveThreshold  ADAPTIVE_TRESH_GAUSSIAN_C

```python
radioactivity = cv.imread(r"radioactivity.jpg",0)
readable_marie_curie = cv.adaptiveThreshold(radioactivity,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)

import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.imshow(radioactivity,'gray')
plt.subplot(1,2,2)
plt.imshow(readable_marie_curie,'gray')
plt.show()
```

<h3>threshold TRESH_OTSU

<h3>erode

```python
import numpy as np

transformation_matrix = np.ones((6,6),np.uint8)

mozaic_pythagoras = cv.erode(statue_colored, transformation_matrix, iterations=2)

plt.imshow(cv.cvtColor(mozaic_pythagoras,cv.COLOR_BGR2RGB))
plt.show()
```

<h3>dilate

```python
import numpy as np

transformation_matrix = np.ones((6,6),np.uint8)

mozaic_pythagoras = cv.dilate(statue_colored, transformation_matrix, iterations=1)

plt.imshow(cv.cvtColor(mozaic_pythagoras,cv.COLOR_BGR2RGB))
plt.show()
```

<h3>VideoCapture

```python
capture = cv.VideoCapture("preview.avi")
while capture.isOpened():
    valid,frame = capture.read()
    if valid:
        cv.imshow("Hydra",frame)
        cv.waitKey(20)
    else:
        break
capture.release()
cv.destroyAllWindows()
```

<img src="hydra.png">

<h3>webcam

```python
webcam = cv.VideoCapture(0)
while webcam.isOpened():
    valid,frame = webcam.read()
    if valid:
        cv.imshow("Hydra",frame)
        if cv.waitKey(1) == ord("q"):
            break
    else:
        break
webcam.release()
cv.destroyAllWindows()
```
